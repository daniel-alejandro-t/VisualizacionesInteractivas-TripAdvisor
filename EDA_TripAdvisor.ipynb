## **Visualizaciones**

### Análisis de distribución de datos

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['avg_rating'].dropna(), bins=20, kde=True)
plt.title('Distribución de las valoraciones')
plt.xlabel('Valoración promedio')
plt.ylabel('Frecuencia')
plt.show()
```

## Matriz de correlación

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df[['avg_rating', 'total_reviews_count', 'reviews_count_in_default_language', 'excellent', 'very_good', 'average', 'poor', 'terrible', 'food', 'service', 'value']].corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
```

#### La matriz de correlación muestra cómo dos variables se relacionan entre sí. Los valores pueden variar entre -1 y 1.

*   1 indica una correlación positiva perfecta (si una variable aumenta, la otra también lo hace).
*   -1 indica una correlación negativa perfecta (si una variable aumenta, la otra disminuye).
*   0 indica que no hay correlación entre las variables.


## Análisis por categorías


### ¿Cuales son los tipos de cocina mas comunes?

Examinamos los tipos de cocina más comunes (cuisines):

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que tu DataFrame es df y ya tiene la columna 'cuisines'
# Convertimos las cadenas de texto en listas
df['cuisines'] = df['cuisines'].dropna().apply(lambda x: x.strip("[]").replace("'", "").split(", "))
all_cuisines = df.explode('cuisines')

# Obtener los 10 tipos de cocina más comunes
top_cuisines = all_cuisines['cuisines'].value_counts().head(10)

# Configurar colores
colors = ['#FF6347', '#4682B4', '#FFD700', '#ADFF2F', '#FF69B4', '#87CEFA', '#32CD32', '#FFA07A', '#20B2AA', '#778899']

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
top_cuisines.plot(kind='bar', color=colors)
plt.title('Tipos de cocina más comunes')
plt.xlabel('Tipo de cocina')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

### ¿Cuales son las 10 Ciudades con mas restaurantes?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
ciudades = df['city'].value_counts().head(10)
ax = ciudades.plot(kind='bar', color='skyblue')  # Puedes cambiar el color si lo deseas

# Mostrar valores numéricos encima de cada barra
for i, v in enumerate(ciudades):
    ax.text(i, v + 10, str(v), ha='center', va='bottom', fontsize=10)

plt.title('Ciudades con más restaurantes')
plt.xlabel('Ciudad')
plt.ylabel('Número de restaurantes')
plt.xticks(rotation=45)
plt.show()
```

### ¿A que país pertenece la mayoría de restaurantes?

```python
import matplotlib.pyplot as plt

# Agrupamos aquellos paises minoritarios en el valor 'Otros países'
frequencies = df['country'].replace(df['country'].value_counts().tail(n=18).index, 'Otros países').value_counts()

# Crea un pie chart de country con los valores agrupados
frequencies.plot.pie(autopct='%1.1f%%')
plt.title('Distribución de restaurantes por país')
plt.ylabel('')
plt.show()
```
